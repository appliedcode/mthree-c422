## Rule-based NLP

```python
# Rule-based example using simple pattern matching
import re

text = "Please schedule a meeting with Prem tomorrow at 9 AM."

if re.search(r'\bplease\b', text.lower()) and re.search(r'\b(schedule|book|arrange)\b', text.lower()):
    print("Detected Intent: Set Meeting")
else:
    print("Intent not recognized.")
```

## ML-based NLP

```python
#ML-based NLP
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
from transformers import pipeline, logging
logging.set_verbosity_error()

# Load zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Input text
text = "Please schedule a meeting with John tomorrow at 10 AM."
labels = ["Set Meeting", "Cancel Meeting", "Weather Query"]

# Predict intent
result = classifier(text, labels)
print("Predicted Intent:", result['labels'][0])
```

```python
## Exercises

1. **Rule-based NLP:**
    Write a rule-based function to detect if a text is asking to cancel a meeting. Test it with the sentence:
    `"Can you please cancel my meeting with Alice at 2 PM?"`

2. **ML-based NLP:**
    Use the zero-shot classification pipeline to predict the intent for the following text:
    `"Will it rain tomorrow in New York?"`
    Use the labels: `["Set Meeting", "Cancel Meeting", "Weather Query"]`.

3. **Custom Labels:**
    Add your own label to the list of intents (e.g., `"Order Food"`) and test the classifier with a relevant sentence.

4. **Compare Approaches:**
    Compare the output of rule-based and ML-based approaches for the text:
    `"Please book a meeting with Sarah next Monday."`
    Which approach gives a more accurate intent prediction?
```

```python
# 1. Rule-based NLP: Function to detect cancelling a meeting
def detect_cancel_meeting(text):
    """
    Detects if a text is asking to cancel a meeting using rule-based pattern matching.

    Args:
        text (str): The input text.

    Returns:
        str: The detected intent or "Intent not recognized."
    """
    if re.search(r'\b(cancel|postpone)\b', text.lower()) and re.search(r'\bmeeting\b', text.lower()):
        return "Detected Intent: Cancel Meeting"
    else:
        return "Intent not recognized."

# Test the function
text_to_test = "Can you please cancel my meeting with Alice at 2 PM?"
print(detect_cancel_meeting(text_to_test))
```

```python
# 2. ML-based NLP: Predict intent for weather query
text_weather = "Will it rain tomorrow in New York?"
labels_weather = ["Set Meeting", "Cancel Meeting", "Weather Query"]

result_weather = classifier(text_weather, labels_weather)
print("Predicted Intent:", result_weather['labels'][0])
```

```python
# 3. Custom Labels: Add a custom label and test the classifier
custom_labels = ["Set Meeting", "Cancel Meeting", "Weather Query", "Order Food"]
text_food_order = "I'd like to order a pizza for dinner."

result_food_order = classifier(text_food_order, custom_labels)
print("Predicted Intent:", result_food_order['labels'][0])
```

```python
# 4. Compare Approaches: Compare rule-based and ML-based for a specific text
text_comparison = "Please book a meeting with Sarah next Monday."

# Define the rule-based function again for this cell's scope
import re
def detect_cancel_meeting(text):
    """
    Detects if a text is asking to cancel a meeting using rule-based pattern matching.

    Args:
        text (str): The input text.

    Returns:
        str: The detected intent or "Intent not recognized."
    """
    if re.search(r'\b(cancel|postpone)\b', text.lower()) and re.search(r'\bmeeting\b', text.lower()):
        return "Detected Intent: Cancel Meeting"
    else:
        return "Intent not recognized."


# Rule-based approach
print("Rule-based approach:")
print(detect_cancel_meeting(text_comparison)) # Using the function created in Exercise 1, which is for 'Cancel Meeting' intent.

# ML-based approach
print("\nML-based approach:")
labels_comparison = ["Set Meeting", "Cancel Meeting", "Weather Query"]
result_comparison = classifier(text_comparison, labels_comparison)
print("Predicted Intent:", result_comparison['labels'][0])

# Comparison analysis
print("\nComparison Analysis:")
print("For the text 'Please book a meeting with Sarah next Monday.':")
print("- The rule-based approach (designed for 'Cancel Meeting') incorrectly identifies the intent as 'Intent not recognized.'")
print("- The ML-based approach correctly predicts the intent as 'Set Meeting'.")
print("In this case, the ML-based approach provides a more accurate intent prediction as it can understand the meaning of 'book a meeting' in context, whereas the rule-based approach relies on specific keywords ('cancel' or 'postpone').")
```
