## Part 2: Core Principles with Solved Examples

### 1. Fairness

Fairness is about ensuring the AI system treats all groups without discrimination.

**Activity: Bias Detection in Dataset**


```
import pandas as pd

# Sample dataset with gender and loan approval status
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
        'Loan_Approved': [1, 0, 0, 1, 1, 1, 0, 1]}

df = pd.DataFrame(data)

# Calculate approval rates by gender
approval_rate = df.groupby('Gender')['Loan_Approved'].mean()
print("Loan Approval Rate by Gender:")
print(approval_rate)
```

**Explanation \& Results:**

- Male approval rate = 1.0 (100%)
- Female approval rate = 0.25 (25%)

This indicates a possible bias favoring males in loan approval decisions, which is unfair.

**Reflect:**
Approval rates are not equal, highlighting bias in the dataset that should be addressed.

### 2. Transparency

Transparency means making AI decisions understandable.

**Activity: Explain a Simple Modelâ€™s Decision**

```
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Sample data (features and labels)
X = [[0, 0], [1, 1]]
y = [0, 1]

clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

# Visualizing the decision tree to enhance transparency
plt.figure(figsize=(6,4))
tree.plot_tree(clf, filled=True)
plt.show()
```

**Explanation:**
The decision tree visualization shows how decisions are made based on feature values, helping users see why a prediction is made.

**Reflect:**
Visualizing the model helps users trust and understand AI by revealing internal logic.

### 4. Privacy

Privacy protects sensitive personal information.

**Activity: Data Masking**

```
# Sample sensitive data
data = {'Name': ['Alice', 'Bob'], 'SSN': ['123-45-6789', '987-65-4321']}
df = pd.DataFrame(data)

# Mask SSN for privacy
df['SSN_Masked'] = df['SSN'].apply(lambda x: 'XXX-XX-' + x[-4:])
print(df)
```

**Explanation:**
Masking hides part of sensitive information, reducing exposure risk.

**Reflect:**
Masking helps protect individualsâ€™ privacy when sharing or processing data.

### 5. Safety and Reliability

AI must operate safely, handling errors to avoid failures.

**Activity: Handling Unexpected Input Gracefully**

```
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero"

print(safe_divide(10, 2))  # Output: 5.0
print(safe_divide(10, 0))  # Output: Error: Division by zero
```

**Explanation:**
The function manages errors gracefully instead of crashing, ensuring reliability.

**Reflect:**
Error handling is critical to keep AI systems robust and safe for users.
