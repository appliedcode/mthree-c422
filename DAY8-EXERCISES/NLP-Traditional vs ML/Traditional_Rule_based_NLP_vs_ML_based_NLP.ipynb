```python
# Rule-based example using simple pattern matching
import re

text = "Please schedule a meeting with Prem tomorrow at 9 AM."

if re.search(r'\bplease\b', text.lower()) and re.search(r'\b(schedule|book|arrange)\b', text.lower()):
    print("Detected Intent: Set Meeting")
else:
    print("Intent not recognized.")
```

```python
#ML-based NLP
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
from transformers import pipeline, logging
logging.set_verbosity_error()

# Load zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Input text
text = "Please schedule a meeting with John tomorrow at 10 AM."
labels = ["Set Meeting", "Cancel Meeting", "Weather Query"]

# Predict intent
result = classifier(text, labels)
print("Predicted Intent:", result['labels'][0])
```

## Exercises

1. **Rule-based NLP:**  
    Write a rule-based function to detect if a text is asking to cancel a meeting. Test it with the sentence:

2. **ML-based NLP:**

```python
# Exercise1

import re

def detect_cancel_meeting(text):
    # Check if the sentence implies canceling a meeting
    cancel_keywords = r'\b(cancel|delete|remove|call off|drop)\b'
    meeting_keywords = r'\b(meeting|appointment|call)\b'

    if re.search(cancel_keywords, text.lower()) and re.search(meeting_keywords, text.lower()):
        print("Detected Intent: Cancel Meeting")
    else:
        print("Intent not recognized.")

# Test input
text1 = "Can you please cancel my meeting with Alice at 2 PM?"
detect_cancel_meeting(text1)
```

```python
# Exe-2
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

from transformers import pipeline, logging
logging.set_verbosity_error()

# Load classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text2 = "Will it rain tomorrow in New York?"
labels = ["Set Meeting", "Cancel Meeting", "Weather Query"]

result = classifier(text2, labels)
print("Predicted Intent:", result['labels'][0])

```

```python
# exercise 3
text3 = "I'd like to order a pizza and a coke for lunch."

custom_labels = ["Set Meeting", "Cancel Meeting", "Weather Query", "Order Food"]

result_custom = classifier(text3, custom_labels)
print("Predicted Intent:", result_custom['labels'][0])
```

```python
#exe-4 Rule based
def detect_set_meeting(text):
    if re.search(r'\b(book|schedule|arrange)\b', text.lower()) and re.search(r'\b(meeting|call|appointment)\b', text.lower()):
        print("Rule-based Prediction: Set Meeting")
    else:
        print("Rule-based Prediction: Intent not recognized.")

text4 = "Please book a meeting with Sarah next Monday."
detect_set_meeting(text4)
```

```python
# ML based
result_compare = classifier(text4, labels)
print("ML-based Prediction:", result_compare['labels'][0])
```
