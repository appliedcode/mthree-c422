````python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Create synthetic dataset for environmental monitoring
np.random.seed(123)
n = 400
districts = ["Industrial Area", "Residential Area", "Parkland", "Commercial Zone"]
pollution_levels = ["Low", "Moderate", "High", "Critical"]
data_env = pd.DataFrame({
    "Sensor_ID": range(1, n + 1),
    "District": np.random.choice(districts, n, p=[0.3, 0.3, 0.2, 0.2]),
    "Air_Quality_Index": np.random.randint(20, 300, n),
    "Water_Contamination_Level": np.random.randint(1, 10, n),
    "AI_Pollution_Alert": np.random.choice([0, 1], n, p=[0.8, 0.2]),
    "Public_Trust_in_AI": np.random.choice(["High", "Medium", "Low"], n, p=[0.3, 0.5, 0.2])
})
````

````python
# Create pollution categories based on Air Quality Index
bins = [0, 50, 100, 150, 300]
labels = pollution_levels
data_env["Air_Quality_Category"] = pd.cut(data_env["Air_Quality_Index"], bins=bins, labels=labels)
# Display the first few rows of the dataset
print(data_env.head())
````

````python
print(data_env.columns.tolist())
````

````python
# Analyze pollution levels across districts
pollution_summary = data_env.groupby("District")["Air_Quality_Index"].value_counts().unstack(fill_value=0)
print("\nPollution levels by district:")
print(pollution_summary)
# Plot pollution levels by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
pollution_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("Pollution Levels by District")
plt.xlabel("District")
plt.ylabel("Number of Sensors")
plt.legend(title="Air Quality Category")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
## Analyze AI pollution alerts by district
alert_summary = data_env.groupby("District")["AI_Pollution_Alert"].value_counts().unstack(fill_value=0)
print("\nAI Pollution Alerts by District:")
print(alert_summary)
# Plot AI pollution alerts by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
alert_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("AI Pollution Alerts by District")
plt.xlabel("District")
plt.ylabel("Number of Sensors")
plt.legend(title="AI Pollution Alert")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze public trust in AI by district
trust_summary = data_env.groupby("District")["Public_Trust_in_AI"].value_counts().unstack(fill_value=0)
print("\nPublic Trust in AI by District:")
print(trust_summary)
# Plot public trust in AI by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
trust_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("Public Trust in AI by District")
plt.xlabel("District")
plt.ylabel("Number of Sensors")
plt.legend(title="Public Trust in AI")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze water contamination levels across districts
water_summary = data_env.groupby("District")["Water_Contamination_Level"].describe()
print("\nWater Contamination Levels Summary by District:")
print(water_summary)
# Plot water contamination levels by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
data_env.boxplot(column="Water_Contamination_Level", by="District", ax=ax)
plt.title("Water Contamination Levels by District")
plt.suptitle("")
plt.xlabel("District")
plt.ylabel("Water Contamination Level")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze air quality categories across districts
air_quality_summary = data_env.groupby("District")["Air_Quality_Category"].value_counts().unstack(fill_value=0)
print("\nAir Quality Categories by District:")
print(air_quality_summary)
# Plot air quality categories by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
air_quality_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("Air Quality Categories by District")
plt.xlabel("District")
plt.ylabel("Number of Sensors")
plt.legend(title="Air Quality Category")
plt.xticks(rotation=45, ha="right")
plt.show()
plt.savefig('air_quality_categories_by_district.png', bbox_inches='tight')
````

````python
# Analyze the relationship between AI pollution alerts and public trust
trust_alert_summary = data_env.groupby("Public_Trust_in_AI")["AI_Pollution_Alert"].value_counts().unstack(fill_value=0)
print("\nAI Pollution Alerts by Public Trust in AI:")
print(trust_alert_summary)
# Plot AI pollution alerts by public trust in AI
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
trust_alert_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("AI Pollution Alerts by Public Trust in AI")
plt.xlabel("Public Trust in AI")
plt.ylabel("Number of Sensors")
plt.legend(title="AI Pollution Alert")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze the relationship between air quality index and public trust
air_quality_trust_summary = data_env.groupby("Public_Trust_in_AI")["Air_Quality_Index"].describe()
print("\nAir Quality Index Summary by Public Trust in AI:")
print(air_quality_trust_summary)
# Plot air quality index by public trust in AI
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
data_env.boxplot(column="Air_Quality_Index", by="Public_Trust_in_AI", ax=ax)
plt.title("Air Quality Index by Public Trust in AI")
plt.suptitle("")
plt.xlabel("Public Trust in AI")
plt.ylabel("Air Quality Index")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze the relationship between water contamination levels and air quality categories
water_air_quality_summary = data_env.groupby("Air_Quality_Category")["Water_Contamination_Level"].describe()
print("\nWater Contamination Levels Summary by Air Quality Category:")
print(water_air_quality_summary)
# Plot water contamination levels by air quality category
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
data_env.boxplot(column="Water_Contamination_Level", by="Air_Quality_Category", ax=ax)
plt.title("Water Contamination Levels by Air Quality Category")
plt.suptitle("")
plt.xlabel("Air Quality Category")
plt.ylabel("Water Contamination Level")
plt.xticks(rotation=45, ha="right")
plt.show()
````

````python
# Analyze the relationship between air quality index and water contamination levels
correlation = data_env["Air_Quality_Index"].corr(data_env["Water_Contamination_Level"])
print(f"\nCorrelation between Air Quality Index and Water Contamination Level: {correlation:.2f}")
# Plot scatter plot of air quality index vs water contamination level
plt.figure(figsize=(10, 6))
plt.scatter(data_env["Air_Quality_Index"], data_env["Water_Contamination_Level"], alpha=0.5)
plt.title("Air Quality Index vs Water Contamination Level")
plt.xlabel("Air Quality Index")
plt.ylabel("Water Contamination Level")
plt.grid()
plt.show()
````

````python
# average public trust vs broken down by district
trust_district_summary = data_env.groupby("District")["Public_Trust_in_AI"].value_counts(normalize=True).unstack(fill_value=0)
print("\nPublic Trust in AI by District (Normalized):")
print(trust_district_summary)
# Plot average public trust in AI by district
fig, ax = plt.subplots(figsize=(14, 8), constrained_layout=True)
trust_district_summary.plot(kind='bar', stacked=True, ax=ax)
plt.title("Average Public Trust in AI by District")
plt.xlabel("District")
plt.ylabel("Proportion of Sensors")
plt.legend(title="Public Trust in AI")
plt.xticks(rotation=45, ha="right")
plt.show()
````
