```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/refs/heads/main/Exercises/day-4/Conversion_techniques/homeprices.csv')
df
```

```python
df_dummies = pd.get_dummies(df['town'], dtype = 'int')
df_dummies
```

```python
df0 = pd.concat([df, df_dummies], axis = 1)
df0
```

```python
df1 = df0.drop(['town', 'west windsor'], axis = 1)
df1
```

```python
X = df1.drop(['price'], axis = 1)
X
```

```python
y = df1['price']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
model.predict(X)
```

```python
model.score(X, y)
```

```python
model.predict([[3400,0,0]])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = df
dfle
```

```python
print(type(dfle))
```

```python
dfle['town'] = le.fit_transform(dfle['town'])
dfle
```

```python
y = df['price']
X = dfle[['town', 'area']].values
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X , y)
```
