```python
from openai import OpenAI
import os
from google.colab import userdata

api_key = userdata.get('OPENAI_API_KEY')
if not api_key:
    raise ValueError("‚ùå API key not found.")
os.environ["OPENAI_API_KEY"] = api_key
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

Exercise 1: Instruction Clarity and Detail

```python
prompt = "Explain how photosynthesis works in three sentences."
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=100
)
print(response.choices[0].message.content)
```

```python
prompt = (
    "Explain how photosynthesis works in detail, using bullet points with specific steps and examples."
)
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=150
)
print(response.choices[0].message.content)
```

Exercise 8: Multi-turn Conversation Simulation

```python
prompt = (
    "Simulate a customer support conversation for a broken printer. Ask me clarifying questions before offering a solution, and confirm my answers before proceeding."
)
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=180
)
print(response.choices[0].message.content)
```

```python
messages = [
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": "Can you tell me the make and model of your printer?"},
    {"role": "user", "content": "model is hp-689"},
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": "<model's second question>"},
    {"role": "user", "content": "is it not turning on"},
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": "Is the printer receiving power? When you press the power button, does any light come on or do you hear any sounds?"},
    {"role": "user", "content": "yes"},
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": "Is this a wired or wireless printer? If it's wireless, is it connected to your Wi-Fi network?"},
    {"role": "user", "content": "It is wireless and yes it is connected to Wi-Fi network"},
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": "Have you made any recent changes to your printer setup or computer, such as new software installations or updates?"},
    {"role": "user", "content": "no"}
    # and so on
]
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    max_tokens=100
)
print(response.choices[0].message.content)
```
