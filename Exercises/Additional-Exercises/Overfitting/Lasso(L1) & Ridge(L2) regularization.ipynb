```python
import pandas as pd
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.drop(['Address', 'Date', 'Lattitude', 'Longtitude', 'YearBuilt', 'Postcode'], axis =1, inplace = True)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[['Bedroom2', 'Bathroom', 'Car']] = df[['Bedroom2', 'Bathroom', 'Car']].fillna(0)
df.head()
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df.Landsize.fillna(df.Landsize.mean())
df['BuildingArea'] = df.BuildingArea.fillna(df.BuildingArea.mean())
df.isna().sum()
```

```python
df.dropna(inplace = True)
df.shape
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
X = df.drop(['Price'], axis = 1)
y = df.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression
li_reg = LinearRegression()
li_reg.fit(X_train, y_train)
li_reg.score(X_test, y_test)
```

```python
li_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(max_iter = 10000)
lasso_reg.fit(X_train, y_train)
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(max_iter = 10000)
ridge_reg.fit(X_train, y_train)
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
