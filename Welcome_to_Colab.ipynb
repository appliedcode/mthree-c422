### Exercise 1: Weather & Climate Data Integration

First, I'll load the historical weather data from the provided CSV file using pandas.

```
import pandas as pd

csv_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/us-weather-history/KCLT.csv"
df_weather_csv = pd.read_csv(csv_url)
display(df_weather_csv.head())
```

Next, I'll load the sample JSON data.

```
import requests
import json

json_url = "https://raw.githubusercontent.com/public-apis/public-apis/master/example.json"
response = requests.get(json_url)
response.raise_for_status()  # Raise an exception for bad status codes

# Load the entire content as a single JSON object
json_data = response.json()

# Displaying a part of the JSON data as it can be quite large
display(json_data)
```

Now, let's fetch current weather data from the Open-Meteo REST API.

```
import requests

api_url = "https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current_weather=true"
response = requests.get(api_url)
response.raise_for_status() # Raise an exception for bad status codes
api_data = response.json()

display(api_data)
```
