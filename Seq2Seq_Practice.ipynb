```python
from transformers import pipeline

# Create a list of English sentences
english_sentences = [
    "I like to read books in the evening.",
    "The weather is sunny today.",
    "We are learning about natural language processing.",
    "She plays the piano very well.",
    "They are going to the park this afternoon."
]

# Load the French translation pipeline
translator = pipeline("translation_en_to_fr")

# Translate each sentence and print the original and translated versions
for sentence in english_sentences:
    translated_sentence = translator(sentence)[0]['translation_text']
    print(f"Original: {sentence}")
    print(f"Translation: {translated_sentence}\n")
```

```python
from transformers import pipeline

# Sample sentence
sample_sentence = "He works at Apple and eats an apple every day."

# Load the English to German translation pipeline
translator_en_de = pipeline("translation_en_to_de")

# Translate to German
translated_sentence_de = translator_en_de(sample_sentence)[0]['translation_text']
print(f"Original: {sample_sentence}")
print(f"German Translation: {translated_sentence_de}\n")

# Load the English to Spanish translation pipeline
translator_en_es = pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")

# Translate to Spanish
translated_sentence_es = translator_en_es(sample_sentence)[0]['translation_text']
print(f"Original: {sample_sentence}")
print(f"Spanish Translation: {translated_sentence_es}\n")
```

```python
from transformers import pipeline

# Load the French to English translation pipeline
translator_fr_en = pipeline("translation_fr_to_en", model="Helsinki-NLP/opus-mt-fr-en")

# French sentence to translate back to English
french_sentence = "J'aime lire des livres le soir."

# Translate the French sentence to English
translated_sentence_en = translator_fr_en(french_sentence)[0]['translation_text']
print(f"Original French: {french_sentence}")
print(f"Translated English: {translated_sentence_en}\n")
```

```python
from transformers import pipeline

def translate_long_text(text, chunk_size=100):
    """
    Translates a long English text into French by splitting it into chunks.

    Args:
        text (str): The long English text to translate.
        chunk_size (int): The approximate size of each chunk in tokens or characters.

    Returns:
        str: The translated French text.
    """
    # Load the French translation pipeline
    translator = pipeline("translation_en_to_fr")

    # Split the text into chunks (simple character-based splitting for demonstration)
    # A more robust approach would use a tokenizer to split by tokens
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

    translated_chunks = []
    for chunk in chunks:
        translated_chunk = translator(chunk)[0]['translation_text']
        translated_chunks.append(translated_chunk)

    # Join the translated chunks
    return " ".join(translated_chunks)

# Example usage with a long paragraph
long_paragraph = """
Natural Language Processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human (natural) languages. As such, NLP is related to the area of human–computer interaction. Many challenges in NLP involve speech recognition, natural language understanding, and natural language generation.

NLP has been around for more than 50 years and has its origins in the field of linguistics. However, it was not until the 1950s that the first attempts at machine translation were made. These early systems were based on simple rule-based approaches and achieved limited success. The 1980s saw the rise of statistical methods in NLP, which led to significant improvements in performance. In recent years, deep learning techniques have further advanced the state of the art in many NLP tasks.
"""

translated_paragraph = translate_long_text(long_paragraph)
print(f"Original Paragraph:\n{long_paragraph}\n")
print(f"Translated Paragraph (chunked):\n{translated_paragraph}")
```

```python
from transformers import pipeline

# Load the French translation pipeline
translator = pipeline("translation_en_to_fr")

# Sentence to translate
sentence = "The quick brown fox jumps over the lazy dog."

# Translate with different max_length values
translated_10 = translator(sentence, max_length=10)[0]['translation_text']
translated_20 = translator(sentence, max_length=20)[0]['translation_text']
translated_40 = translator(sentence, max_length=40)[0]['translation_text']

print(f"Original: {sentence}\n")
print(f"Translation (max_length=10): {translated_10}")
print(f"Translation (max_length=20): {translated_20}")
print(f"Translation (max_length=40): {translated_40}")
```
