```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

from transformers import pipeline, logging

# Suppress unnecessary logs from transformers
logging.set_verbosity_error()
```

```python
def get_translator(src_lang="en", tgt_lang="fr"):
    """
    Returns a translation pipeline for the given source and target languages.
    """
    model_name = f"Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}"
    try:
        translator_pipeline = pipeline(f"translation_{src_lang}_to_{tgt_lang}", model=model_name)
        return translator_pipeline
    except Exception as e:
        print(f"Error loading model '{model_name}': {e}")
        return None
```

```python
def translate_sentences(sentences, src_lang="en", tgt_lang="fr", max_length=50):
    translator_pipeline = get_translator(src_lang, tgt_lang)
    if not translator_pipeline:
        return []

    results = translator_pipeline(sentences, max_length=max_length)
    return [res['translation_text'] for res in results]
```

```python
if __name__ == "__main__":
    # Example: multiple sentences to translate
    texts = [
        "He works at Apple and eats an apple every day.",
        "The weather is beautiful today.",
        "I love programming in Python."
    ]

    # Choose source and target languages
    source_lang = "en"  # English
    target_lang = "fr"  # French

    # Translate
    translations = translate_sentences(texts, src_lang=source_lang, tgt_lang=target_lang)
```

```python
for original, translated in zip(texts, translations):
        print(f"Original ({source_lang}): {original}")
        print(f"Translated ({target_lang}): {translated}\n")
```
