```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/homeprices.csv')
df
df.head()
df.tail()
```

```python
model = LinearRegression()
model.fit(df[['area']], df[['price']])
```

```python
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
sns.scatterplot(x='area', y='price', data=df, color='blue', s=100, label='Training Data')
sns.lineplot(x=df['area'], y=model.predict(df[['area']]).flatten(), color='red', label='Regression Line')
plt.title('House Price vs Area (Training Data)', fontsize=16, fontweight='bold')
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
areas_df = pd.read_csv('/content/areas.csv')
df
```

```python
areas_df['predicted_price'] = model.predict(areas_df[['area']])
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='area', y='predicted_price', hue='area', data=areas_df, palette='viridis', legend=False)
plt.title('Predicted House Prices for New Areas', fontsize=16, fontweight='bold')
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Predicted Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
output_path = '/content/predictions.csv'
areas_df.to_csv(output_path, index=False)
```
