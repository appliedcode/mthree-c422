```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
mi_scores = mutual_info_classif(X, y, random_state=42)
df_mi = pd.DataFrame({
    'feature': X.columns,
    'MI_score': mi_scores
}).sort_values('MI_score', ascending=False)

print("Mutual Information scores:\n", df_mi)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='MI_score', y='feature', data=df_mi)
plt.title('Mutual Information Scores')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)
anova_df = pd.DataFrame({'feature': X.columns, 'F_score': selector.scores_})
```

```python
merged = anova_df.merge(df_mi, on='feature')
print("\nComparison of F-scores and MI scores:")
print(merged.sort_values(['F_score', 'MI_score'], ascending=False))
```

```python
top_features = df_mi['feature'][:3]
for feat in top_features:
    plt.figure()
    sns.stripplot(x=y, y=X[feat], jitter=0.2)  # stripplot supports jitter
    plt.title(f"{feat} vs. Target")
    plt.xlabel("Target")
    plt.ylabel(feat)
    plt.tight_layout()
    plt.show()
```

```python
correlation = merged[['F_score', 'MI_score']].corr().iloc[0, 1]
print(f"\nCorrelation between F-scores and MI scores: {correlation:.4f}")
```
