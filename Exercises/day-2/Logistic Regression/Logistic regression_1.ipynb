```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df.sample(5)
```

```python
df.describe()
```

```python
df_1 = df.copy()
df_1['salary'] = df_1['salary'].replace({'low': 1, 'medium': 3, 'high': 5}) 
df_1.sample(5)
```

```python
dummies = pd.get_dummies(df['Department'], dtype = 'int')
dummies
```

```python
df_2 = df_1.drop(['Department'], axis = 1)
df_2.sample(5)
```

```python
df_2.groupby('left').mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (5,4))
sns.barplot(x = 'salary', y = 'left', data = df_1)
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.barplot(x = 'Department', y = 'left', data = df)
plt.show()
```

```python
dummies_1 = pd.get_dummies(df['salary'], dtype = 'int')
dummies_1.head()
```

```python
df_2 = pd.concat([df_2, dummies_1], axis = 1)
df_2 = df_2.drop(['salary'], axis = 1)
df_2.sample(5)
```

```python
max(df_2['average_montly_hours'])
```

```python
y = df_2['left']
X = df_2.drop(['left'], axis = 1)
X['average_montly_hours'] = X['average_montly_hours']/310
X.sample(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 5, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
