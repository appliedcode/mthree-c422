## Part 1: Impact of AI on Jobs and Economic Inequality

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Synthetic dataset: Job automation risk vs wage distribution
np.random.seed(0)
n = 500

data_jobs = pd.DataFrame({
    "Occupation": [f"Job_{i}" for i in range(1, n+1)],
    "Automation_Risk": np.random.beta(2, 5, n),  # Risk score between 0 (low) and 1 (high)
    "Median_Wage": np.random.normal(50000, 15000, n)  # USD annual wages
})

# Visualize relationship between automation risk and wages
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Automation_Risk', y='Median_Wage', data=data_jobs)
plt.title('Job Automation Risk vs Median Wage')
plt.xlabel('Automation Risk (0=Low, 1=High)')
plt.ylabel('Median Annual Wage (USD)')
plt.show()
```
