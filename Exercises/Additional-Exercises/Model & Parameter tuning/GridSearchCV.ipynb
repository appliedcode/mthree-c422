```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df.sample(8)
```

```python
y = iris.target
X = df.drop(['flower'], axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
from sklearn.svm import SVC

C = [1,10,40]
kernels = ['rbf', 'linear']
avg_scores = {}

for i in kernels:
    for j in C:
        cv_scores = cross_val_score(SVC(kernel = i, C = j, gamma = 'auto'), X, y, cv = 5)
        avg_scores[i + '_' + str(j)] = np.average(cv_scores)

avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma = 'auto'), {'kernel': ['rbf', 'linear'], 'C': [1, 10, 40]}, cv = 5, return_train_score = False)
clf.fit(X, y)
```

```python
dir(clf)
```

```python
clf.cv_results_
```

```python
results = pd.DataFrame(clf.cv_results_)
results_1 = results[['params', 'mean_test_score']]
results_1
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm' :{
        'model': svm.SVC(gamma = 'auto'),
        'parameters': {
            'kernel' : ['rbf', 'linear'],
            'C': [1, 10, 40]}},

    'lr':{
        'model': LogisticRegression(solver='liblinear'),
        'parameters': {
            'C': [1,5,10]}},
    'rf': {
        'model': RandomForestClassifier(),
        'parameters': {
            'n_estimators': [1,5,10]}}
}


```

```python
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['parameters'], cv = 5, return_train_score = False)
    clf.fit(X, y)
    scores.append({
        'model': model_name,
        'best_params': clf.best_params_,
        'score': clf.best_score_})
```

```python
scores
```

```python
scores_df = pd.DataFrame(scores, columns= ['model', 'best_params', 'score'])
scores_df
```
