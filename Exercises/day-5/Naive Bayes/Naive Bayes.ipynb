```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/refs/heads/main/Exercises/day-5/Naive%20Bayes/titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 1, inplace =True)
df.head()
```

```python
df.columns[df.isna().any()]
```

```python
df['Age'].head(10)
```

```python
df['Age'] = df['Age'].fillna(df.Age.mean())
df.head(10)
```

```python
y = df['Survived']
X = df.drop(['Survived'], axis = 1)
X.head(6)
```

```python
X['Sex'] = X['Sex'].replace({'male':1 , 'female': 0})
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, X_train, y_train, cv = 8)
```
