```python
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
import pandas as pd
```

```python
df  = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/refs/heads/main/Exercises/day-5/Principal%20Component%20Analysis/heart.csv')
df.head()
```

```python
df.shape
```

```python
import numpy as np
from scipy.stats import zscore

df_numeric = df.select_dtypes(include = [np.number])

z_scores = np.abs(zscore(df_numeric))

threshold = 3

filtered = (z_scores < threshold).all(axis = 1)

df_1 = df[filtered]
df_1.head()
```

```python
df_1.shape
```

```python
df_2 = pd.get_dummies(df_1, dtype = 'int')
df_2.head()
```

```python
X = df_2.drop('HeartDisease', axis = 1)
y = df_2['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size =0.2, random_state = 5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

models = {'Logistic Regression': LogisticRegression(), 'Random Forest': RandomForestClassifier(), 'SVC': SVC()}


def get_scores(X_train, X_test, y_train, y_test):
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        print(f'{model_name}: {model.score(X_test, y_test)}')
get_scores(X_train, X_test, y_train, y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.75)
X_pca = pca.fit_transform(X_scaled)

X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size =0.2, random_state = 5)
```

```python
get_scores(X_train_pca, X_test_pca, y_train, y_test)
```
