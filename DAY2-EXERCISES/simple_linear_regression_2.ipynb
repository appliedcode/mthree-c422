```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot of the data")
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Slope (coefficient):", lin_reg.coef_[0][0])
print("Intercept:", lin_reg.intercept_[0])
```

```python
# 4. Make Predictions
X_new = np.array([[0], [2]])
y_pred = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, color="red", linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Best Fit Line")
plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/main/Exercises/day-2/Simple%20Linear%20Regression/homeprices.csv', low_memory=False)
print(df.head())
```

```python
X = df[['area']]  # 2D array
y = df[['price']]
```

```python
plt.scatter(X, y)
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.title("Home Prices vs Area")
plt.show()
```

```python
lin_reg.fit(X, y)

print("Slope (coefficient):", lin_reg.coef_[0][0])
print("Intercept:", lin_reg.intercept_[0])
```

```python
X_new = np.array([[1000], [5000]])  # example new areas
y_pred = lin_reg.predict(X_new)

# Plot original data and best fit line
plt.scatter(X, y)
plt.plot(X_new, y_pred, color='red', linewidth=2)
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.title("Linear Regression: Home Price Prediction")
plt.show()
```
