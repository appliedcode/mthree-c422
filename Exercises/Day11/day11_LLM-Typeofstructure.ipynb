```
# Create client
from openai import OpenAI
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

```
# Exercise 1 – Instructions Pattern
prompt = "You are a baker. Give step-by-step instructions (max 6 steps) for baking a cake."

resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "system", "content": "You are a helpful assistant."},
              {"role": "user", "content": prompt}],
    max_tokens=500
)
print(resp.choices[0].message.content)


```

```
# Exercise 2 – Examples (Few‑Shot) Pattern
few_shot = """Translate English → Kannada.

English: Hello !
Kannada: Namaskara !

English: Where is you home?
Kannada: Nimma mane yelli ?

English: My name is Rania.
Kannada:"""

resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": few_shot}],
    max_tokens=50
)
print(resp.choices[0].message.content)
```

```
# Exercise 3 – Persona Pattern
prompt = "You are Ada Lovelace writing a 120-word diary entry about today’s AI advancements."
resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=200
)
print(resp.choices[0].message.content)
```

```
# Exercise 4 – Recipe + Alternative Approaches Pattern
prompt = """Give me a recipe to prepare mojito and ways to serve it"""
resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=100
)
print(resp.choices[0].message.content)
```

```
# Exercise 5 – Ask-for-Input / Flipped Interaction Pattern
prompt = """You are an astrologer. Ask me 3 questions you need for predictions."""
resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=300
)
print(resp.choices[0].message.content)
```

```
# Exercise 6 – Menu / Meta-Language Pattern
menu_prompt = """### MENU
ACTION = Explain, Summarize, Translate
FORMAT = Plain, Bullets

Command: ACTION=Summarize; FORMAT=Bullets;
TEXT="AI will boom in the coming years"
###"""

resp = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": menu_prompt}],
    max_tokens=150
)
print(resp.choices[0].message.content)
```
