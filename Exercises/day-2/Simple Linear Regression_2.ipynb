# Objective
Fit a straight line to a set of data points and use the model to make predictions.

```python
# 1. Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 2. Generate Sample Data
# Create some random data with a linear relationship
np.random.seed(42)
X = 4 * np.random.rand(50, 1)
y = 3 + 3 * X + np.random.randn(50, 1)

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot of the data")
plt.show()
```

```python
# 3. Create and Fit the Model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Slope (coefficient):", lin_reg.coef_[0][0])
print("Intercept:", lin_reg.intercept_[0])
```

```python
# 4. Make Predictions
X_new = np.array([[0], [2]])
y_pred = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, color="red", linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Best Fit Line")
plt.show()
```
