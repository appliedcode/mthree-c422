```python

# 1. Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 2. Generate Sample Data
# Create some random data with a linear relationship
np.random.seed(500)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Scatter plot of the data")
plt.show()

```

```python

# 3. Create and Fit the Model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Slope (coefficient):", lin_reg.coef_[0][0])
print("Intercept:", lin_reg.intercept_[0])
```

```python
# 4. Make Predictions
X_new = np.array([[0], [2]])
y_pred = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, color="green", linewidth=10)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Best Fit Line")
plt.show()
```

**Using the Pandas**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(1)
X = 3 * np.random.rand(100, 1)
y = 2 + 5 * X + np.random.randn(100, 1)

```

```python
df = pd.read_csv("predictions.csv")
X = df[["area"]]
y = df[["prices"]]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
print("Slope (coefficient):", lin_reg.coef_[0][0])
print("Intercept:", lin_reg.intercept_[0])
```

```python
X_new = np.array([[min(X.values)[0]], [max(X.values)[0]]])
y_pred = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, color="black", linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Best Fit Line from CSV Data")
plt.grid(True)
plt.show()
```
