```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# --- PROMPT / INPUT ---
texts = [
    "The prime minister met foreign leaders to discuss trade agreements.",
    "The local team won the championship after a thrilling final.",
    "Stock markets surged after tech companies posted strong earnings."
]
candidate_labels = ["sports", "politics", "business"]
# -------------------------------

for text in texts:
    print("=== PROMPT SENT TO MODEL ===")
    print(text)
    result = classifier(text, candidate_labels)
    print(f"Predicted Label: {result['labels'][0]} (Score: {result['scores'][0]:.2f})")
    print("-" * 60)
```

```python
generator = pipeline("text-generation", model="tiiuae/falcon-7b-instruct", device_map="auto")

# --- THIS IS THE PROMPT ---
prompt = """Classify the sentiment of the following movie reviews as Positive or Negative.

Review: "I loved this film! The story was captivating."
Sentiment: Positive

Review: "This was the worst acting I have ever seen."
Sentiment: Negative

Review: "Absolutely wonderful soundtrack and direction."
Sentiment: Positive

Review: "The plot was dull and uninteresting."
Sentiment:"""
# --------------------------

print("\n=== PROMPT SENT TO MODEL ===")
print(prompt)

response = generator(prompt, max_new_tokens=20, temperature=1.0)

print("\n=== MODEL RESPONSE ===")
print(response[0]['generated_text'])
```

```python
# Install once in Colab if needed:
# !pip install -q transformers torch

import torch
from transformers import pipeline

# choose device for pipeline: 0 for GPU, -1 for CPU
device = 0 if torch.cuda.is_available() else -1

# Small fast model for local testing (no API key needed)
generator = pipeline("text-generation", model="distilgpt2", device=device)

# IMPORTANT: prompt must be a Python string (use triple quotes for multiline)
prompt = """Q: Tom has 3 apples. He buys 2 more packs with 4 apples each. Then he eats 3 apples. How many apples does he have left?
Let's think step-by-step.

1. Tom starts with 3 apples.
2. He buys 2 packs of 4 apples = 2 * 4 = 8 apples.
3. Total apples now = 3 + 8 = 11 apples.
4. He eats 3 apples, so remaining = 11 - 3 = 8 apples.
Answer: 8

Q: Sarah has 10 pencils. She gives 3 to John and 2 to Mary. How many does she have left?
Let's think step-by-step.

1. Sarah starts with 10 pencils.
2. She gives 3 pencils to John => 10 - 3 = 7
3. She gives 2 pencils to Mary => 7 - 2 = 5
4. She is left with 5 pencils.
Answer:"""

print("\n=== PROMPT SENT TO MODEL ===\n")
print(prompt)

# Generate (deterministic: temperature=0)
response = generator(prompt, max_new_tokens=40, temperature=0, do_sample=False)

generated = response[0]["generated_text"]
print("\n=== MODEL RESPONSE (raw) ===\n")
print(generated)

# Optionally extract only the model's continuation after the prompt:
continuation = generated[len(prompt):].strip()
print("\n=== MODEL CONTINUATION (after prompt) ===\n")
print(continuation)

# Try to extract a short 'Answer:' line if present
if "Answer:" in generated:
    answer = generated.rsplit("Answer:", 1)[-1].strip().split("\n")[0]
    print("\n=== Extracted Answer ===\n")
    print(answer)
```
