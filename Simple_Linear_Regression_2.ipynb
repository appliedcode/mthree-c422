```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(10)
X = 3 * np.random.rand(100, 1)  # Change scale from 2 to 3
y = 2 + 5 * X + 2 * np.random.randn(100, 1)  # Change intercept to 2 and slope to 5, more noise

# Plot new data
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Modified Random Data")
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("New Slope (coefficient):", lin_reg.coef_[0][0])
print("New Intercept:", lin_reg.intercept_[0])
```

```python
X_new = np.array([[0], [3]])
y_pred = lin_reg.predict(X_new)

plt.scatter(X, y)
plt.plot(X_new, y_pred, color="green", linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Modified Best Fit Line")
plt.show()

```
