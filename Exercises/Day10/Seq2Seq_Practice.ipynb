Exercise 1: Translate Multiple Sentences

```python
from transformers import pipeline

# English to French translator
en_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")

sentences = [
    "I wake up at 7 am every day.",
    "She enjoys painting landscapes on weekends.",
    "Technology has changed the way we live.",
    "They play football every evening after work.",
    "My favorite food is pasta with cheese."
]

for sentence in sentences:
    translation = en_fr(sentence)[0]['translation_text']
    print(f"EN: {sentence}\nFR: {translation}\n")
```

Exercise 2: Experiment with Different Translation Models

```python
# English to German and Spanish
en_de = pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de")
en_es = pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")

sample = "He works at Apple and eats an apple every day."

de_translation = en_de(sample)[0]['translation_text']
es_translation = en_es(sample)[0]['translation_text']

print(f"Original: {sample}")
print(f"German: {de_translation}")
print(f"Spanish: {es_translation}")
```

Exercise 3: Reverse Translation (French to English)

```python
# Reverse translator: French to English
fr_en = pipeline("translation_fr_to_en", model="Helsinki-NLP/opus-mt-fr-en")

# Example: Use French sentence from Exercise 1
french_sent = en_fr("I wake up at 7 am every day.")[0]['translation_text']
back_to_english = fr_en(french_sent)[0]['translation_text']

print(f"Original English: I wake up at 7 am every day.")
print(f"French: {french_sent}")
print(f"Back Translated English: {back_to_english}")
```

Exercise 4: Handle Long Text Translation

```python
def split_text(text, max_chars=100):
    return [text[i:i+max_chars] for i in range(0, len(text), max_chars)]

def translate_long_text(text):
    chunks = split_text(text)
    translated_chunks = [en_fr(chunk)[0]['translation_text'] for chunk in chunks]
    return " ".join(translated_chunks)

long_paragraph = """
Artificial Intelligence is rapidly transforming industries around the world. From healthcare and education to transportation and finance, AI systems are being integrated to increase efficiency and improve user experience.
In healthcare, for example, AI algorithms are used to diagnose diseases, recommend treatments, and predict patient outcomes.
In education, intelligent tutoring systems personalize learning experiences for students, helping them achieve better results.
Meanwhile, autonomous vehicles are reshaping transportation by reducing human error and improving road safety.
In finance, AI-driven algorithms analyze market trends, manage portfolios, and detect fraudulent activities.
Despite the advantages, the widespread adoption of AI raises important questions about data privacy, bias in algorithms, and the potential for job displacement.
It is essential for policymakers, developers, and society at large to work together to ensure AI technologies are developed responsibly and ethically.
"""

translated_paragraph = translate_long_text(long_paragraph)
print(f"\nTranslated Long Paragraph (French):\n{translated_paragraph}")
```

Exercise 5: Translation with Controlled Output Length

```python
sentence = "The quick brown fox jumps over the lazy dog."

print("Original:", sentence)
for max_len in [10, 20, 40]:
    translated = en_fr(sentence, max_length=max_len)[0]['translation_text']
    print(f"max_length={max_len} → {translated}")
```
