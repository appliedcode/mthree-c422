```
from openai import OpenAI
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])  # Replace with your API key

def ask_gpt(prompt, temp=0.7, max_tokens=500):
    """Send a prompt to GPT-4o-mini and return the output"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=temp,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content
```

## **Principle 1 ‚Äî Give Clear Direction**

**Goal:** Be specific about *what*, *for whom*, and *how* the task should be done.

```
prompt = "Write a 50-word friendly welcome message for first-year university students in an encouraging tone."
print(ask_gpt(prompt))
```

```
# TODO: Your prompt here
vague_prompt = "Write about energy."
clear_prompt = "Explain renewable energy sources (solar, wind, hydropower) to high school students in an informative but simple tone, 100 words or less."

print("Vague Output:\n", ask_gpt(vague_prompt))
print("\nClear Output:\n", ask_gpt(clear_prompt))
```

## **Principle 2 ‚Äî Specify Format**

**Goal:** Tell the AI exactly how to structure its response.

```
para_prompt = "Compare Phone A, Phone B, and Phone C."
table_prompt = "Compare Phone A, Phone B, and Phone C in a markdown table with columns: Phone, Display, Camera, Battery."

print("Paragraph Output:\n", ask_gpt(para_prompt))
print("\nTable Output:\n", ask_gpt(table_prompt))
```

## **Principle 4 ‚Äî Evaluate Quality**

**Goal:** Review the AI output and refine the prompt to improve quality.


```
text = """Community gardens are flourishing in cities worldwide, providing fresh produce and green spaces for residents. They improve air quality, foster social connections, and promote sustainability."""
# Initial vague prompt
vague_sum = ask_gpt(f"Summarize this text: {text}")

# Improved prompt
refined_sum = ask_gpt(f"Summarize this article in 3 bullet points each covering: 1) Main benefits 2) Social impact 3) Environmental effects : {text}")

print("Original Summary:\n", vague_sum)
print("\nRefined Summary:\n", refined_sum)
```

## **Principle 5 ‚Äî Divide Labor (Break Down Complex Tasks)**


```
# Step 1
audience_analysis = ask_gpt("Describe the target audience for an online bookstore focused on rare and antique books.")
# Step 2
social_media_strategy = ask_gpt("Based on the above audience, create a marketing strategy with 5 bullet points.")
# Step 3
content_calendar = ask_gpt("Propose a 1-month social media content calendar in a table with Date, Post Idea, and Caption.")

print("Audience Analysis:\n", audience_analysis)
print("\nSocial Media Strategy:\n", social_media_strategy)
print("\nContent Calendar:\n", content_calendar)
```
